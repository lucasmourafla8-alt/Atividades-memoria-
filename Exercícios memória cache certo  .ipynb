Exemplo 5.2 — Mapeamento Direto (Total de Bits)
Enunciado: Considere um sistema de computação com uma memória cache de 32KB de capacidade, constituída de linhas com 8 bytes de largura. A MP possui uma capacidade de 16MB. Calcule o total de bits da cache.
Resolução:
$$T = T_d + T_t$$
1. Bits de dados (Td): $$T_d = \text{tamanho da cache} \times 8 = 32\text{KB} \times 8 = 2^{18} \text{ bits}$$
2. Número de linhas da cache (N): $$N = \frac{32\text{KB}}{8\text{B}} = 2^{12} \text{ linhas}$$
3. Bits de tag:
$$\text{bits_tag} = \text{bits_MP} - \text{bits_linha} - \text{bits_byte}$$
MP = 16MB = $2^{24}$ → 24 bits
Linhas = $2^{12}$ → 12 bits
Byte = 8B = $2^{3}$ → 3 bits
$$\text{bits_tag} = 24 - 12 - 3 = 9 \text{ bits}$$
4. Bits de tag total (Tt): $$T_t = N \times \text{bits_tag} = 2^{12} \times 9$$
5. Total da cache: $$T = 2^{18} + 2^{12} \times 9 = 262.144 + 36.864 = \textbf{299.008 bits}$$
Exemplo 5.3 — Mapeamento Direto (Formato do Endereço)
Enunciado: Considere uma MP com 64MB de capacidade associada a uma memória cache que possui 2K linhas, cada uma com largura de 16 bytes. Determine o formato do endereço.
Resolução:
1. Bits de byte: $$\text{Byte} = 16\text{B} = 2^{4} \Rightarrow \textbf{4 bits}$$
2. Bits de linha: $$\text{Linhas} = 2\text{K} = 2^{11} \Rightarrow \textbf{11 bits}$$
3. Bits de tag: $$\text{bits_tag} = \text{bits_MP} - \text{bits_linha} - \text{bits_byte}$$ $$\text{MP} = 64\text{MB} = 2^{26} \Rightarrow 26 \text{ bits}$$ $$\text{bits_tag} = 26 - 11 - 4 = \textbf{11 bits}$$
Formato do Endereço:

Linha
Byte
11 bits
11 bits
4 bits
$$\text{Total} = 11 + 11 + 4 = \textbf{26 bits}$$

Exemplo 5.5 — Associativo Completo (Total de Bits)
Enunciado: Considere um sistema com cache de 32KB, linhas de 8 bytes e MP de 16MB. Calcule o total de bits da cache.
Resolução:
$$T = T_d + T_t$$
1. Bits de dados (Td): $$T_d = 32\text{KB} \times 8 = 2^{18} \text{ bits}$$
2. Número de linhas da cache (N): $$N = \frac{32\text{KB}}{8\text{B}} = 2^{12} \text{ linhas}$$
3. Bits de tag:
No associativo completo não existe campo linha, então: $$\text{bits_tag} = \text{bits_MP} - \text{bits_byte}$$ $$\text{MP} = 16\text{MB} = 2^{24} \Rightarrow 24 \text{ bits}$$ $$\text{Byte} = 8\text{B} = 2^{3} \Rightarrow 3 \text{ bits}$$ $$\text{bits_tag} = 24 - 3 = \textbf{21 bits}$$
4. Total: $$T = 2^{18} + 2^{12} \times 21 = 262.144 + 86.016 = \textbf{348.160 bits}$$

3. Bits de bloco/tag: $$\text{bits_bloco} = \text{bits_MP} - \text{bits_byte} = 28 - 5 = \textbf{23 bits}$$

Exemplo 5.8 — Associativo por Conjunto (Total de Bits)
Enunciado: Considere um sistema com cache de 32KB, linhas de 8 bytes e conjunto de 4. A MP possui 16MB. Calcule o total de bits da cache.
Resolução:
$$T = T_d + T_t$$
1. Bits de dados (Td): $$T_d = 32\text{KB} \times 8 = 2^{18} \text{ bits}$$
2. Número de linhas da cache (N): $$N = \frac{32\text{KB}}{8\text{B}} = 2^{12} \text{ linhas}$$
3. Número de conjuntos: $$\text{Conjuntos} = \frac{\text{linhas da cache}}{\text{linhas por conjunto}} = \frac{2^{12}}{2^{2}} = 2^{10} \Rightarrow \textbf{10 bits}$$
4. Bits de tag: $$\text{bits_tag} = \text{bits_MP} - \text{bits_conjunto} - \text{bits_byte}$$ $$\text{MP} = 16\text{MB} = 2^{24} \Rightarrow 24 \text{ bits}$$ $$\text{bits_tag} = 24 - 10 - 3 = \textbf{11 bits}$$
5. Total: $$T = 2^{18} + 2^{12} \times 11 = 262.144 + 45.056 = \textbf{307.200 bits}$$
