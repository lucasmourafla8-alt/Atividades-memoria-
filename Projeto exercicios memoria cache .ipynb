## Exemplo 5.2 - Cálculo de total de bits para Cache com Mapeamento Direto

### Enunciado
Considere um sistema de computação com uma memória cache de 32KB de capacidade, constituída de linhas com 8 bytes de largura. A MP possui uma capacidade de 16MB.

### Resolução Teórica
1. **Total de bits de dados ($T_d$):**
   $$T_d = 32\text{ KB} \times 8\text{ bits} = 32 \times 1024 \times 8 = 262.144\text{ bits}$$

2. **Quantidade de linhas da cache ($L$):**
   $$L = \frac{32\text{ KB}}{8\text{ B}} = 4\text{ K} = 2^{12}\text{ linhas}$$

3. **Quantidade de blocos da MP:**
   $$\text{Blocos} = \frac{16\text{ MB}}{8\text{ B}} = 2\text{ MB} = 2^{21}\text{ blocos}$$

4. **Quantidade de blocos por linha (Largura do campo Tag):**
   $$\text{Blocos por linha} = \frac{2^{21}}{2^{12}} = 2^9 = 512 \implies \text{Tag} = 9\text{ bits}$$

5. **Total de bits de tag ($T_t$):**
   $$T_t = 4\text{ K linhas} \times 9\text{ bits} = 4 \times 1024 \times 9 = 36.864\text{ bits}$$

6. **Capacidade Total da Cache ($T$):**
   $$T = T_d + T_t = 262.144 + 36.864 = 299.008\text{ bits} \approx 292\text{ KBits}$$

# Validando Exemplo 5.2
cache_cap = 32 * 1024 # em Bytes
linha_largura = 8 # em Bytes
mp_cap = 16 * 1024 * 1024 # em Bytes

t_d = cache_cap * 8
linhas = cache_cap / linha_largura
blocos_mp = mp_cap / linha_largura
tag_bits = 21 - 12 # log2(blocos_mp) - log2(linhas)
t_t = linhas * tag_bits
t_total = t_d + t_t

print(f"=== EXEMPLO 5.2 ===")
print(f"Bits de Dados (Td): {t_d} bits")
print(f"Bits de Tag (Tt): {int(t_t)} bits")
print(f"Total da Cache: {int(t_total)} bits ({t_total/1024:.1f} KBits)")

## Exemplo 5.3 - Formato de Endereço (Mapeamento Direto)

### Enunciado
Considere uma MP com 64MB de capacidade associada a uma memória cache que possui 2K linhas, cada uma com largura de 16 bytes. Determine o formato do endereço.

### Resolução Teórica
* **Capacidade da MP:** $64\text{ MB} = 2^{26}\text{ bytes} \implies \text{Endereço total} = 26\text{ bits}$
* **Largura do bloco/linha:** $16\text{ B} = 2^4\text{ bytes} \implies \text{Campo Byte} = 4\text{ bits}$
* **Linhas da Cache ($L$):** $2\text{ K} = 2^{11}\text{ linhas} \implies \text{Campo Linha} = 11\text{ bits}$
* **Campo Tag:**
   $$\text{Tag} = \text{Total} - \text{Linha} - \text{Byte} = 26 - 11 - 4 = 11\text{ bits}$$

**Formato do Endereço:** [ Tag: 11 bits | Linha: 11 bits | Byte: 4 bits ] = 26 bits

# Validando Exemplo 5.3
mp_bits = 26 # log2(64M)
byte_bits = 4 # log2(16)
linha_bits = 11 # log2(2K)
tag_bits = mp_bits - linha_bits - byte_bits

print(f"=== EXEMPLO 5.3 ===")
print(f"Campos do Endereço -> Tag: {tag_bits} bits | Linha: {linha_bits} bits | Byte: {byte_bits} bits")
print(f"Total: {tag_bits + linha_bits + byte_bits} bits")

## Exemplo 5.4 - Mapeamento de Endereço Hexadecimal (Mapeamento Direto)

### Enunciado
Seja uma MP constituída de blocos com largura de 32 bytes, associada a uma cache com 128KB. O processador realiza um acesso no endereço hexadecimal `3FC92B6`. Determine o valor binário da linha.

### Resolução Teórica
1. **Endereço em bits:** `3FC92B6` possui 7 dígitos hexadecimais $\implies 7 \times 4 = 28\text{ bits}$.
2. **Campo Byte:** Largura de 32 bytes $= 2^5 \implies 5\text{ bits}$.
3. **Linhas da Cache:** $\frac{128\text{ KB}}{32\text{ B}} = 4\text{ K} = 2^{12}\text{ linhas} \implies 12\text{ bits}$.
4. **Campo Tag:** $28 - 12 - 5 = 11\text{ bits}$.

Convertendo `3FC92B6` para binário e separando nos campos [Tag: 11 | Linha: 12 | Byte: 5]:
* Binário total: `0011 1111 1100 1001 0010 1011 0110`
* Divisão: `[00111111110] [010010010101] [10110]`

**Resultado:** O endereço da linha procurada em binário é `010010010101`.

# Validando Exemplo 5.4
endereco_hex = "3FC92B6"
endereco_bin = bin(int(endereco_hex, 16))[2:].zfill(28)

# Divisão dos campos: Tag(11), Linha(12), Byte(5)
tag = endereco_bin[0:11]
linha = endereco_bin[11:23]
byte = endereco_bin[23:28]

print(f"=== EXEMPLO 5.4 ===")
print(f"Endereço Binário Completo: {endereco_bin}")
print(f"Campo Tag  (11 bits): {tag}")
print(f"Campo Linha(12 bits): {linha} <--- Resposta")
print(f"Campo Byte (5 bits) : {byte}")

## Exemplo 5.5 - Mapeamento Associativo Completo (Total de Bits)

### Enunciado
Considere um sistema com cache de 32KB, linhas de 8 bytes e MP de 16MB. Calcule os bits da cache usando Mapeamento Associativo Completo.

### Resolução Teórica
1. **Bits de dados ($T_d$):** $32\text{ KB} \times 8 = 262.144\text{ bits}$.
2. **Linhas da cache:** $\frac{32\text{ KB}}{8\text{ B}} = 4\text{ K} = 2^{12}\text{ linhas}$.
3. **Tamanho do campo Bloco (Tag):** No associativo completo, a tag armazena o endereço do bloco inteiro.
   $$\text{Total de Blocos da MP} = \frac{16\text{ MB}}{8\text{ B}} = 2\text{ MB} = 2^{21} \implies \text{Campo Bloco} = 21\text{ bits}$$
4. **Bits de Bloco/Tag ($T_b$):** $4\text{ K linhas} \times 21\text{ bits} = 86.016\text{ bits}$.
5. **Total da Cache:** $T = 262.144 + 86.016 = 348.160\text{ bits} \approx 340\text{ KBits}$.

## Exemplo 5.8 - Mapeamento Associativo por Conjunto (Total de Bits)

### Enunciado
Cache de 32KB, linhas de 8 bytes, organização por **Conjunto de 4** (Four-way set associative). MP de 16MB.

### Resolução Teórica
1. **Bits de dados ($T_d$):** $32\text{ KB} \times 8 = 262.144\text{ bits}$.
2. **Linhas da cache:** $\frac{32\text{ KB}}{8\text{ B}} = 4\text{ K} = 2^{12}\text{ linhas}$.
3. **Quantidade de Conjuntos ($C$):** Como cada conjunto tem 4 linhas:
   $$C = \frac{4\text{ K linhas}}{4} = 1\text{ K conjuntos} = 2^{10}\text{ conjuntos}$$
4. **Quantidade de blocos da MP:** $\frac{16\text{ MB}}{8\text{ B}} = 2\text{ MB} = 2^{21}\text{ blocos}$.
5. **Tamanho da Tag:** $\frac{\text{Blocos}}{\text{Conjuntos}} = \frac{2^{21}}{2^{10}} = 2^{11} \implies \text{Tag} = 11\text{ bits}$.
6. **Bits de Tag ($T_t$):** $4\text{ K linhas} \times 11\text{ bits} = 45.056\text{ bits}$ (Nota: o livro calculou $T_t = 4\text{K} \times 2\text{K}$ erroneamente na imagem impressa como $8192$, o valor corrigido por linha com tag de 11 bits é $4\text{K} \times 11 = 45.056$).
7. **Total real corrigido:** $262.144 + 45.056 = 307.200\text{ bits}$.

## Exemplo 5.9 - Formato de Endereço (Associativo por Conjunto)

### Enunciado
MP com 64MB ($2^{26}$), cache de 32KB com blocos de 16 bytes ($2^4$), mapeamento associativo por conjunto de 4.

### Resolução Teórica
1. **Campo Byte:** Bloco de 16B $= 2^4 \implies 4\text{ bits}$.
2. **Linhas da cache:** $\frac{32\text{ KB}}{16\text{ B}} = 8\text{ K} = 2^{13}\text{ linhas}$.
3. **Conjuntos:** $\frac{8\text{ K}}{4} = 2\text{ K conjuntos} = 2^{11} \implies \text{Campo Conjunto} = 11\text{ bits}$.
4. **Campo Tag:** $\text{Total} - \text{Conjunto} - \text{Byte} = 26 - 11 - 4 = 11\text{ bits}$.

**Formato final do endereço:** [ Tag: 11 bits | Conjunto: 11 bits | Byte: 4 bits ]
