# Simulador Memória Tannenbaum 128K x 4 bits

class MemoriaTannenbaum:
    def __init__(self):
        # 128K posições = 131072 endereços
        # cada posição guarda 4 bits
        self.memoria = [0] * (128 * 1024)

    def acessar(self, linha, coluna):
        # junta linha e coluna para formar endereço
        endereco = (linha << 8) | coluna
        
        return endereco


    def escrever(self, linha, coluna, dado):
        # CS ativo e WE ativo
        endereco = self.acessar(linha, coluna)

        if dado < 0 or dado > 15:
            print("Erro: dado deve ter 4 bits")
            return

        self.memoria[endereco] = dado
        print(f"Escrito {dado:04b} no endereço {endereco}")


    def ler(self, linha, coluna):
        # CS ativo e OE ativo
        endereco = self.acessar(linha, coluna)

        dado = self.memoria[endereco]

        print(f"Lido {dado:04b} do endereço {endereco}")

        return dado



# Programa principal

ram = MemoriaTannenbaum()


# Exemplo de escrita
ram.escrever(
    linha=10,
    coluna=5,
    dado=9
)


# Exemplo de leitura
ram.ler(
    linha=10,
    coluna=5
)
