class MemoriaTannenbaum:

    def __init__(self):

        # 2 bancos
        # cada banco tem 128K posições de 4 bits
        self.banco0 = [0] * (128 * 1024)
        self.banco1 = [0] * (128 * 1024)


    def acessar(self, linha, coluna):

        endereco = (linha << 8) | coluna

        return endereco


    def escrever(self, banco, linha, coluna, dado):

        endereco = self.acessar(linha, coluna)


        if dado < 0 or dado > 15:
            print("Dado inválido")
            return


        if banco == 0:
            self.banco0[endereco] = dado

        elif banco == 1:
            self.banco1[endereco] = dado


        print(f"Banco {banco} endereço {endereco}: dado {dado:04b}")


    def ler(self, banco, linha, coluna):

        endereco = self.acessar(linha, coluna)


        if banco == 0:
            dado = self.banco0[endereco]

        else:
            dado = self.banco1[endereco]


        print(f"Leitura Banco {banco}: {dado:04b}")

        return dado



# Criando memória
ram = MemoriaTannenbaum()


# Escrevendo no Banco 0
ram.escrever(0, 10, 5, 9)


# Lendo do Banco 0
ram.ler(0, 10, 5)



# Escrevendo no Banco 1
ram.escrever(1, 20, 3, 7)


# Lendo do Banco 1
ram.ler(1, 20, 3)
