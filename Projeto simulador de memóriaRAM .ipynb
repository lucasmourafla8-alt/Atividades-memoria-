# Simulador de memória RAM

# Criando a memória com 16 posições, todas iniciando com 00000000
memoria = ["00000000"] * 16


# Converter endereço binário para número
def converter_endereco(endereco):
    return int(endereco, 2)


# Escrever um dado na memória
def escrever():
    endereco = input("Digite o endereço (0000 a 1111): ")
    dado = input("Digite o dado de 8 bits: ")

    posicao = converter_endereco(endereco)

    memoria[posicao] = dado

    print("Dado gravado com sucesso!")


# Ler um dado da memória
def ler():
    endereco = input("Digite o endereço que deseja ler: ")

    posicao = converter_endereco(endereco)

    print("Valor armazenado:", memoria[posicao])


# Mostrar toda a memória
def listar():
    print("\nMemória RAM:")

    for i in range(16):
        print(format(i, "04b"), "->", memoria[i])


# Menu principal
while True:

    print("\n1 - Escrever")
    print("2 - Ler")
    print("3 - Listar memória")
    print("4 - Sair")

    opcao = input("Escolha uma opção: ")


    if opcao == "1":
        escrever()

    elif opcao == "2":
        ler()

    elif opcao == "3":
        listar()

    elif opcao == "4":
        print("Programa encerrado")
        break

    else:
        print("Opção inválida")
