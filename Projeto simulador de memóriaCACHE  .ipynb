# Simulador de Memória Cache

# Memória RAM com 32 posições
ram = []

for i in range(32):
    ram.append(format(i, "08b"))


# Cache com 4 linhas
cache = [
    {"tag": "", "byte": "", "dado": "00000000"},
    {"tag": "", "byte": "", "dado": "00000000"},
    {"tag": "", "byte": "", "dado": "00000000"},
    {"tag": "", "byte": "", "dado": "00000000"}
]


# Separar endereço em TAG, LINHA e BYTE

def separar_endereco(endereco):

    tag = endereco[0:2]

    linha = int(endereco[2:4], 2)

    byte = endereco[4]

    return tag, linha, byte



# Ler memória

def ler():

    endereco = input("Digite endereço de 5 bits: ")

    tag, linha, byte = separar_endereco(endereco)


    print("\nConsultando cache...")


    # Comparador de TAG

    if cache[linha]["tag"] == tag:

        print("CACHE HIT (acerto)")
        print("Dado:", cache[linha]["dado"])


    else:

        print("CACHE MISS (falha)")
        print("Buscando na RAM...")


        posicao = int(endereco,2)

        dado = ram[posicao]


        cache[linha]["tag"] = tag
        cache[linha]["byte"] = byte
        cache[linha]["dado"] = dado


        print("Cache atualizada")
        print("Dado:", dado)



# Mostrar cache

def mostrar():

    print("\nCACHE")

    print("Linha | TAG | BYTE | DADO")


    for i in range(4):

        print(
            i,
            "|",
            cache[i]["tag"],
            "|",
            cache[i]["byte"],
            "|",
            cache[i]["dado"]
        )



# Menu

while True:

    print("\n1 - Ler endereço")
    print("2 - Mostrar cache")
    print("3 - Sair")


    op = input("Escolha: ")


    if op == "1":

        ler()


    elif op == "2":

        mostrar()


    elif op == "3":

        break


    else:

        print("Opção inválida")
