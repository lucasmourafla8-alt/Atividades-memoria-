# Simulador Memória Tannenbaum 512K x 8 bits

class MemoriaTannenbaum:

    def __init__(self):

        # 512K posições
        # Cada posição guarda 8 bits
        self.memoria = [0] * (512 * 1024)



    # RAS = linha
    # CAS = coluna
    def acessar(self, linha, coluna):

        endereco = (linha << 9) | coluna

        return endereco



    # Escrita (WE)
    def escrever(self, linha, coluna, dado):

        endereco = self.acessar(linha, coluna)


        # 8 bits = valores de 0 até 255
        if dado < 0 or dado > 255:
            print("Erro: dado deve ter 8 bits")
            return


        self.memoria[endereco] = dado


        print(
            f"Escrita -> Endereço {endereco} | "
            f"Dado {dado:08b}"
        )



    # Leitura (OE)
    def ler(self, linha, coluna):

        endereco = self.acessar(linha, coluna)


        dado = self.memoria[endereco]


        print(
            f"Leitura -> Endereço {endereco} | "
            f"Dado {dado:08b}"
        )


        return dado




# Criando a memória
ram = MemoriaTannenbaum()


# Gravando um valor
ram.escrever(
    linha=20,
    coluna=10,
    dado=150
)


# Lendo o valor
ram.ler(
    linha=20,
    coluna=10
)
