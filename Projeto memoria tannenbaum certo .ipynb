# Projeto 1 - Sistema de Memória de 8 Gbits

## 1. Definições e Restrições do Enunciado
* Capacidade Total = 8 Gbits = $8 \times 2^{30}$ bits = $2^{33}$ bits.
* Pinos de Endereço do Chip = 13 pinos (multiplexados em RAS/CAS).
* Barramento de Dados (BD) = Transporta 2 palavras por vez.
* Restrições: $BD \le 64$ bits e $\text{Palavra} > \text{Célula}$.

## 2. Processo de Projeto e Justificativas
1. **Tamanho do BD e da Palavra:** Adotando o limite máximo de 64 bits para o BD para maximizar o desempenho. Como ele transporta 2 palavras:
   $$\text{Palavra} = \frac{64}{2} = 32 \text{ bits}$$
2. **Tamanho da Célula:** Como a Palavra (32 bits) deve ser maior que a célula, definimos a célula como potência de 2 comercial padrão:
   $$\text{Célula} = 8 \text{ bits (1 Byte)}$$
3. **Capacidade do Chip:** Utilizando endereçamento multiplexado (13 pinos para linha e 13 para coluna):
   $$\text{Matriz Interna} = 2^{13} \times 2^{13} = 2^{26} \text{ posições}$$
   $$\text{Capacidade de 1 Chip} = 2^{26} \times 8 \text{ bits} = 2^{26} \times 2^3 = 2^{29} \text{ bits (512 Mbits)}$$
4. **Associação de Chips:**
   $$\text{Total de Chips} = \frac{2^{33} \text{ bits}}{2^{29} \text{ bits}} = 2^4 = 16 \text{ chips}$$
5. **Barramento de Endereços do Sistema:** O sistema acessa 64 bits ($2^6$ bits) por vez através do BD:
   $$\text{Posições do Sistema} = \frac{2^{33}}{2^6} = 2^{27}$$
   $$\text{Barramento de Endereços} = 27 \text{ linhas}$$

Total de Chips necessários: {int(total_chips)} chips
""")

# Projeto 2 - Sistema de Memória de 32 Gbits

## 1. Definições e Restrições do Enunciado
* Capacidade Total = 32 Gbits = $32 \times 2^{30}$ bits = $2^{35}$ bits.
* Linhas de Dados do Sistema (BD) = 16 linhas (16 bits).
* Restrição: Pinos de endereço do chip < 14 (Máximo de 13 pinos).

## 2. Processo de Projeto e Justificativas
1. **Largura de Saída (Célula do Chip):** Adotamos chips com organização de 4 bits de saída ($\times4$).
   * *Justificativa:* Como o BD do sistema é de 16 bits, podemos associar perfeitamente 4 chips em paralelo ($4 \times 4 \text{ bits} = 16 \text{ bits}$) para preencher o barramento de dados.
2. **Endereçamento Interno do Chip:** Adotamos o limite de 13 pinos de endereço por chip (atendendo à restrição de ser $< 14$). O endereçamento será multiplexado (13 bits para RAS e 13 bits para CAS).
3. **Capacidade do Chip:**
   $$\text{Matriz Interna} = 2^{13} \times 2^{13} = 2^{26} \text{ posições}$$
   $$\text{Capacidade de 1 Chip} = 2^{26} \times 4 \text{ bits} = 2^{26} \times 2^2 = 2^{28} \text{ bits (256 Mbits)}$$
4. **Associação de Chips e Bancos:**
   $$\text{Total de Chips} = \frac{2^{35} \text{ bits}}{2^{28} \text{ bits}} = 2^7 = 128 \text{ chips}$$
   $$\text{Número de Bancos} = \frac{128 \text{ chips}}{4 \text{ chips por banco}} = 32 \text{ bancos}$$
5. **Barramento de Endereços do Sistema:** O sistema acessa 16 bits ($2^4$ bits) por vez através do BD:
   $$\text{Posições do Sistema} = \frac{2^{35}}{2^4} = 2^{31}$$
   $$\text{Barramento de Endereços} = 31 \text{ linhas}$$
   *(Onde 26 bits vão para a multiplexação dos chips e 5 bits vão para o decodificador de 32 linhas para selecionar o banco via Chip Select)*
