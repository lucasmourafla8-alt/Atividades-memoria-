# Simulador Memória Tannenbaum 128K x 4 bits
# 2 bancos: Banco 0 e Banco 1

class MemoriaTannenbaum:

    def __init__(self):

        # Cada banco possui 64K células de 4 bits
        self.banco0 = [0] * (64 * 1024)
        self.banco1 = [0] * (64 * 1024)


    # RAS seleciona linha e CAS seleciona coluna
    def acessar(self, linha, coluna):

        endereco = (linha << 8) | coluna

        return endereco


    # Escrita na memória
    def escrever(self, banco, linha, coluna, dado):

        endereco = self.acessar(linha, coluna)


        # Memória é de 4 bits (0 até 15)
        if dado < 0 or dado > 15:
            print("Erro: dado deve ter 4 bits")
            return


        # Seleção do banco
        if banco == 0:
            self.banco0[endereco] = dado

        elif banco == 1:
            self.banco1[endereco] = dado

        else:
            print("Banco inexistente")
            return


        print(
            f"Escrita -> Banco {banco} | "
            f"Endereço {endereco} | "
            f"Dado {dado:04b}"
        )



    # Leitura da memória
    def ler(self, banco, linha, coluna):

        endereco = self.acessar(linha, coluna)


        if banco == 0:
            dado = self.banco0[endereco]

        elif banco == 1:
            dado = self.banco1[endereco]

        else:
            print("Banco inexistente")
            return


        print(
            f"Leitura -> Banco {banco} | "
            f"Endereço {endereco} | "
            f"Dado {dado:04b}"
        )

        return dado



# Criando a memória
ram = MemoriaTannenbaum()


# Teste de escrita no Banco 0

)


# Teste de leitura no Banco 0
ram.ler(
