### Questão 1: Total de Bytes de uma MP
*Dado: Barramento de endereços = 12 bits, Palavra = 16 bits (em cada célula).*

* **Explicação:** Se o barramento de endereços tem 12 bits, a memória consegue ter $2^{12} = 4096$ células diferentes. Como cada célula armazena uma palavra de 16 bits (que equivalem a 2 bytes, já que $16 / 8 = 2$), multiplicamos as células pelo tamanho em bytes.
* **Cálculo:** $4096 \text{ células} \times 2 \text{ bytes} =$ **8192 bytes** (ou exatamente 8 KB).

### Questão 4.1: Tamanho de Endereço (Caderno)
*Dado: RAM (MP) com espaço máximo de endereçamento de 2K. Cada célula armazena 24 bits.*

* **Explicação:** Para encontrar o tamanho do endereço, precisamos de transformar a quantidade de posições de memória numa potência de 2. Como $2\text{K} = 2^1 \cdot 2^{10} = 2^{11}$, o expoente diz-nos diretamente quantos bits/fios são necessários para formar todas as combinações.
* **Resultado:** **11 bits**.

### Questão 4.2: Endereço e Células Utilizadas (Caderno)
*Dado: MP com capacidade total de 256 Kbits. Cada célula armazena 8 bits.*

* **Explicação:** Primeiro, descobrimos o número de células ($n$) dividindo a capacidade total pelo tamanho de cada célula: $\frac{256\text{ Kbits}}{8\text{ bits}} = 32\text{K células}$. Convertendo para potência de 2, $32\text{K} = 2^5 \cdot 2^{10} = 2^{15}$ células. O expoente do número de células indica o tamanho do barramento de endereço.
* **Resultado:** Total de células = **$2^{15}$ (ou 32.768 células)**; Tamanho do endereço = **15 bits**.

### Questão 4.3: Parâmetros de REM e RDM (Caderno)
*Dado: MP com capacidade de 2K palavras de 16 bits cada.*

* **Explicação:** * O RDM tem sempre o mesmo tamanho da palavra que é lida da célula, logo **RDM = 16 bits**.
  * A memória tem $2\text{K} = 2^{11}$ palavras, o que significa que precisamos de um endereço de **11 bits** (tamanho do REM).
  * O maior endereço em decimal é o total de posições menos 1 (porque começa no 0), ou seja, $2048 - 1 =$ **2047**.
  * O total de bits é a multiplicação das posições pelo tamanho: $2048 \times 16 =$ **32.768 bits** (ou $2^{15}$ bits).

### Questão 6: Funções do REM e RDM
* **REM (Registrador de Endereços de Memória):** Guarda temporariamente a localização (o índice ou endereço) na memória onde o processador deseja efetuar uma operação de leitura ou de escrita.
* **RDM (Registrador de Dados de Memória):** Guarda temporariamente o conteúdo (o dado real ou instrução) que acabou de ser extraído da memória ou que está prestes a ser guardado nela.

### Questão 7: Barramentos entre Processador e Memória Principal
1. **Barramento de Endereços:** Transporta a indicação de qual célula vai ser acedida. É **unidirecional** (Processador $\rightarrow$ Memória).
2. **Barramento de Dados:** Transporta a informação/dado real que está a ser lido ou escrito. É **bidirecional** (Processador $\leftrightarrow$ Memória).
3. **Barramento de Controle:** Transporta os sinais de comando e sincronismo (indica se a operação é de LEITURA ou ESCRITA). É **bidirecional**.

### Questão 8: Passo a Passo da Operação de Leitura
1. O processador coloca o endereço da célula que pretende ler no registrador **REM**.
2. Esse endereço é enviado através do **Barramento de Endereços** até à Memória Principal (MP).
3. A unidade de controlo envia um sinal de **LEITURA** pelo Barramento de Controle.
4. A MP localiza a célula correspondente, extrai o dado e coloca-o no **Barramento de Dados**.
5. O dado viaja até ao processador e é armazenado no registrador **RDM**.

### Questão 9: Passo a Passo da Operação de Escrita
1. O processador coloca o endereço de destino no **REM** e o dado que deseja guardar no **RDM**.
2. O endereço é enviado pelo **Barramento de Endereços** e o dado é enviado pelo **Barramento de Dados**.
3. A unidade de controlo emite um sinal de **ESCRITA** através do Barramento de Controle.
4. A MP recebe a informação e grava o dado exatamente na célula de memória especificada pelo endereço.

### Questão 10: Capacidade com REM e RDM
*Dado: RDM = 16 bits, REM = 20 bits, célula = 8 bits.*

* **a) Tamanho do barramento de endereços:** Tem exatamente o mesmo tamanho do REM, ou seja, **20 bits**.
* **b) Células lidas em uma única operação:** O RDM transporta 16 bits de cada vez. Como cada célula armazena 8 bits, faz-se $16 / 8 =$ **2 células**.
* **c) Total de bits na MP:** Um REM de 20 bits permite criar $2^{20} = 1.048.576$ endereços diferentes. Multiplicando pelo tamanho de cada célula: $1.048.576 \times 8 \text{ bits} =$ **8.388.608 bits** (ou 1 MB / 8 Mbits).

### Questão 11: Sistema com 32K Células
*Dado: Memória com 32K posições ($32 \times 1024 = 32.768$ células), cada uma com 8 bits.*

* **a) Maior endereço em decimal:** Como os endereços começam a contar no 0, o último índice será o total menos 1: $32.768 - 1 =$ **32.767**.
* **b) Tamanho do barramento de endereços:** Como $32\text{K} = 2^5 \cdot 2^{10} = 2^{15}$, são necessários **15 bits**.
* **c) Bits no REM e RDM:** O **REM tem 15 bits** (tamanho do endereço) e o **RDM tem 8 bits** (tamanho da célula).
* **d) Total máximo de bits na memória:** $32.768 \text{ células} \times 8 \text{ bits} =$ **262.144 bits**.

### Questão 12: Análise com Valores Hexadecimais
*Dado: Endereço = 2C81 (Hexa), Conteúdo = F5A (Hexa).*

* **a) Tamanho do REM e RDM:** Cada dígito hexadecimal corresponde a exatamente 4 bits no sistema binário.
  * O endereço "2C81" tem 4 dígitos $\rightarrow 4 \times 4 =$ **REM com 16 bits**.
  * O conteúdo "F5A" tem 3 dígitos $\rightarrow 3 \times 4 =$ **RDM com 12 bits**.
* **b) Máxima quantidade de bits na memória:** Com um REM de 16 bits, a memória pode ter até $2^{16} = 65.536$ posições. Multiplicando pelo tamanho da célula (12 bits): $65.536 \times 12 =$ **786.432 bits**.

### Questão 13: Localização do caractere "1.º J" em Hexadecimal
*Dado: Endereço inicial = 27FA (Hexa). Grupos de 128 caracteres (As, Bs, Cs...). Cada caractere ocupa 1 byte.*

* **Explicação:** Queremos descobrir onde fica a primeira letra 'J'. Na ordem alfabética, o 'J' é a 10.ª letra. Isto significa que, antes de chegar ao grupo dos 'Js', o sistema preencheu 9 grupos completos de outras letras (de 'A' até 'I').
* **Cálculo de deslocamento:** $9 \text{ grupos} \times 128 \text{ caracteres} = 1152 \text{ bytes}$.
* Convertendo 1152 para Hexadecimal:
  * $1152 / 16 = 72$ (resto 0)
  * $72 / 16 = 4$ (resto 8)
  * Logo, 1152 em decimal é igual a $480$ em Hexadecimal.
* **Endereço Final:** Somamos o deslocamento ao endereço inicial: $27\text{FA} + 480 = \mathbf{2C7A}$.

### Questão 14: Análise de Custo SRAM vs DRAM
*Dado: Interface DRAM = R$ 5,00. Interface SRAM = R$ 1,00. Bit SRAM = R$ 0,00002. Bit DRAM = R$ 0,00001.*

* **Explicação:** Montamos uma desigualdade para descobrir a partir de quantos bits ($X$) a memória DRAM se torna mais económica que a SRAM, compensando o custo mais alto da sua interface.
  $$\text{Custo SRAM} > \text{Custo DRAM}$$
  $$1 + 0,00002 \cdot X > 5 + 0,00001 \cdot X$$
  $$0,00001 \cdot X > 4$$
  $$X > \frac{4}{0,00001} \rightarrow X > 400.000\text{ bits}$$
* **Resultado:** A memória dinâmica (DRAM) passa a ser a opção mais barata quando o sistema necessita de uma capacidade superior a **400.000 bits**.

### Questão 21: Tempo Médio de Acesso com Memória Cache
*Dado: Tempo Cache = 8 ns, Tempo MP = 70 ns, Eficiência (Hit Rate) = 96% (0,96).*

* **Explicação:** Em 96% dos acessos, o processador encontra o dado diretamente na Cache (gasta 8 ns). Nos 4% restantes (misses), ele gasta o tempo de procurar na cache (8 ns) mais o tempo de ir buscar à memória principal (70 ns).
* **Cálculo:** $$T_{médio} = (0,96 \cdot 8) + (0,04 \cdot (8 + 70))$$
  $$T_{médio} = 7,68 + (0,04 \cdot 78) = 7,68 + 3,12 = \mathbf{10,8\text{ ns}}$$

### Questão 26: Projeto Lógico de uma ROM (8 células de 4 bits)
* **Explicação:** Para selecionar 8 células distintas de memória ($8 = 2^3$), o circuito necessita de um decodificador com **3 linhas de entrada de endereço** ($A_0, A_1, A_2$). Este decodificador terá 8 linhas de saída, onde cada uma ativa uma célula específica. Como cada célula armazena uma palavra de 4 bits, o circuito final terá **4 linhas de saída de dados** ($D_0, D_1, D_2, D_3$) conectadas à matriz da memória.

### Questão 27 e 28: Bits Necessários para Endereçamento
* **Questão 27 (Para uma memória de 128G):** Como $128 = 2^7$ e $\text{G (Giga)} = 2^{30}$, temos $2^7 \cdot 2^{30} = 2^{37}$. São necessários **37 bits** no barramento.
* **Questão 28 (Para uma memória de 32K):** Como $32 = 2^5$ e $\text{K (Kilo)} = 2^{10}$, temos $2^5 \cdot 2^{10} = 2^{15}$. São necessários **15 bits** no barramento.

### Questão 31: Armazenamento de Imagem na Memória
*Dado: Resolução = 640 x 420 pontos (píxeis), onde cada ponto ocupa 4 bytes.*

* **a) Memória em bytes para 1 imagem:** Multiplicamos o total de pontos pelo tamanho de cada um: $640 \times 420 \times 4 =$ **1.075.200 bytes**.
* **b) Memória em megabytes para 10 imagens:** Multiplicamos o valor anterior por 10, obtendo $10.752.000 \text{ bytes}$. Dividindo por $1.000.000$ (conversão comercial decimal), resulta em **10,75 MB**.
* **c) Quantas imagens cabem em 128MB de RAM?** Convertendo 128 MB para bytes binários: $128 \times 1024 \times 1024 = 134.217.728 \text{ bytes}$. Dividindo pelo tamanho de uma imagem: $134.217.728 / 1.075.200 = 124,82$. Cabem **124 imagens** completas.

### Questão 32: Capacidade de uma ROM
*Dado: 16 linhas de endereçamento e 4 linhas de saída de dados.*

* **Explicação:** 16 linhas de endereço permitem criar $2^{16} = 65.536$ posições de memória. Como cada posição entrega 4 bits na saída, o total é $65.536 \times 4 = 262.144 \text{ bits}$. Dividindo por 8 para converter em bytes, obtemos **32.768 bytes** (ou 32 KB).

### Questão 34: Análise sobre Expansão de RAM e Velocidade de Discos
* **Resposta:** **Falso com ressalvas**. Adicionar mais memória RAM não altera o limite físico ou a velocidade mecânica de leitura de um disco rígido ou disquete. Contudo, a afirmação tem uma lógica prática: com mais RAM disponível, o sistema operacional consegue criar estruturas de *cache* maiores em memória rápida, evitando ter de ler o disco lento repetidamente. Isto melhora significativamente o desempenho geral percebido pelo utilizador.
